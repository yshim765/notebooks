```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import optuna

import itertools
```

```python
X, Y = np.meshgrid(np.arange(-100, 101), np.arange(-100, 101))

def abs_add(X, Y):
    return np.abs(X) + np.abs(Y)

fig = plt.figure(figsize=[8, 8])

plt.scatter(X, Y, c = abs_add(X, Y))
plt.title("squared")

plt.show()

def ab_add_sin_times_abs(X, Y):
    return abs(X, Y) + np.abs(np.sin(X)*X + np.sin(Y)*Y)

fig = plt.figure(figsize=[8, 8])

plt.scatter(X, Y, c = ab_add_sin_times_abs(X, Y))
plt.title("ab_add_sin_times_abs")

plt.show()

def four_holls(X, Y):
    return ((X > 0) * (Y > 0))*(np.abs(X-50) + np.abs(Y-50)) +\
            ((X <= 0) * (Y > 0))*(np.abs(X+50) + np.abs(Y-50) + 10) +\
            ((X > 0) * (Y <= 0))*(np.abs(X-50) + np.abs(Y+50) + 15) +\
            ((X <= 0) * (Y <= 0))*(np.abs(X+50) + np.abs(Y+50) + 5)

fig = plt.figure(figsize=[8, 8])

plt.scatter(X, Y, c = four_holls(X, Y))
plt.title("four_holls")

plt.show()
```

```python
def objective(trial):
    x = trial.suggest_int('x', -100, 100)
    y = trial.suggest_int('y', -100, 100)
    return four_holls(x, y)

study = optuna.create_study()
study.optimize(objective, n_trials=100)

study.best_params
```

```python
def objective(trial):
    x = trial.suggest_uniform('x', -100, 100)
    y = trial.suggest_uniform('y', -100, 100)
    return four_holls(x, y)

study = optuna.create_study()
study.optimize(objective, n_trials=100)

study.best_params
```
